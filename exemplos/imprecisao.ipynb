# Imprecisão

Para exemplificar a diferença entre $a/b$ e $a * (1/b)$, considere a matriz
$$
A = \begin{bmatrix}
  1 & 2 \\ 3 & 4
\end{bmatrix}
$$
e o escalar $p = 0.987654321$.

```python
from numpy import matrix

A = matrix([[1e3,2e3],[3e3,4e3]])
p = matrix([[0.987654321]])
```

Apliquemos 500 vezes a operação de divisão.

```python
B, C = A.copy(), A.copy()
difs = []
for i in range(500):
  B = B/p
  C = C*(1/p[0,0])
  difs.append(B - C)
```

Capturando a norma infinito das diferenças, podemos observar que as sucessivas iterações levam ao crescimento do erro:

```python
from numpy.linalg import norm

norma_difs = [norm(D) for D in difs]

import matplotlib.pyplot as plt
plt.plot(norma_difs)
plt.xlabel("Passo")
plt.ylabel("$||B-C||_\infty$")
plt.show()
```
